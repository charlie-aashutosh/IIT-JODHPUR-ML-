```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/heart.csv")
```

```python
data.head()
```

```python
x=data.drop(['target'],axis=1) #independent variable
y=data['target'] #dependent variable
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2) #20 percent for testing
```

Using PCA for dimentionality reduction

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=6) #most important 6 features
```

```python
x_reduced_train=pd.DataFrame(pca.fit_transform(x_train))
x_reduced_test=pd.DataFrame(pca.fit_transform(x_test))
```
